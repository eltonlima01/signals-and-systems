1. Considere o sinal de tempo contínuo definido pela expressão $x(t) = (t^{2} − 1) * [u(t - 1) - u(t - 2)]$.

**a)** Plote o sinal $x(t)$ para $0 \leq t \leq 3$.

```python
import numpy
import matplotlib.pyplot as plt

t1, t2 = 0, 3
points = 1000
t = numpy.linspace (t1, t2, points)

u1 = 1.0 * (t >= 1)
u2 = 1.0 * (t >= 2)
x = ((t**2) - 1) * (u1 - u2)

plt.figure (figsize = (7, 3))
plt.plot (t, x, "k")
plt.xlabel ("$t$")
plt.ylabel ("$x(t) = (t^{2} - 1) [u(t - 1) - u(t - 2)]$")
plt.xlim ([t1, t2])
plt.grid ()
plt.show ()
```

**b)** Calcule, de forma analítica, a energia do sinal $x(t)$.

$E_{x} = \int_{-\infty}^{+\infty} x^{2}(t)dt$

$E_{x} = \int_{1}^{2} \{(t^{2} - 1)[u(t - 1) - u(t - 2)]\}^{2}dt = \int_{1}^{2} (t^{2} − 1)^{2}dt$

$E_{x} = (\frac{t^{5}}{5} - \frac{2t^{3}}{3} + t)\mid_{1}^{2} = \frac{32}{5} - \frac{2 * 8}{3} + 2 - (\frac{1}{5} - \frac{2}{3} + 1)$

$E_{x} \approx 2.53333333333333$

**c)** Determine, de forma numérica, a energia do sinal $x(t)$. Compare com o resultado do item anterior.

```python
print ("Energia (computacional) =", (t2 - t1) / points * numpy.sum (x**2))
print ("Energia (analítica)     =", 2.53333333333333)
```

2. Considere o sinal $x(t)$ de tempo contínuo dado conforme a Figura 1 abaixo.

**a)** Determine, de forma analítica, uma expressão matemática para o sinal $x(t)$ para todo $t$.

$x(t) = 2, -1.5 \leq t \leq 0$\
$x(t) = 2e^{-\frac{t}{2}}, 0 \leq t \leq 3$\
$x(t) = 0,$ caso contrário

$x(t) = (2)[u(t + 1.5) - u(t)] + (2e^{-\frac{t}{2}})[u(t) - u(t - 3)]$

**b)** Plote o sinal $x(t)$.

```python
t1, t2 = -2.5, 4
t = numpy.linspace (t1, t2, points)

u1 = 1.0 * (t >= -1.5)
u2 = 1.0 * (t >= 0)
x1 = (2) * (u1 - u2)

u1 = 1.0 * (t >= 0)
u2 = 1.0 * (t >= 3)
x2 = (2 * numpy.exp (-t/2)) * (u1 - u2)

x = x1 + x2

plt.figure (figsize = (7, 3))
plt.plot (t, x, "k")
plt.xlabel ("$t$")
plt.ylabel ("$x(t)$")
plt.xlim ([t1, t2])
plt.grid ()
plt.show ()
```

**b)** Plote os seguintes sinais:

- (i) $x(-t)$
- (ii) $x(t + 4)$
- (iii) $x(t - 2)$
- (iv) $x(2t)$
- (v) $x(\frac{t}{2})$

```python
t1, t2 = -8, 8
t = numpy.linspace (t1, t2, 2000)

def plot (time):
    u1 = 1.0 * (time >= -1.5)
    u2 = 1.0 * (time >= 0)
    x1 = (2) * (u1 - u2)

    u1 = 1.0 * (time >= 0)
    u2 = 1.0 * (time >= 3)
    x2 = (2 * numpy.exp (-time/2)) * (u1 - u2)

    x = x1 + x2

    plt.figure (figsize = (7, 3))
    plt.plot (t, x, "k")
    plt.xlabel ("$t$")
    plt.ylabel ("$x(t)$")
    plt.xlim ([t1, t2])
    plt.grid ()
    plt.show ()

print ("x(-t)")
plot (-t)

print ("x(t + 4)")
plot (t + 4)

print ("x(t - 2)")
plot (t - 2)

print ("x(2t)")
plot (2 * t)

print ("x(t/2)")
plot (t/2)
```
